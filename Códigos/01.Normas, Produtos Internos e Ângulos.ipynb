# Bibliotecas e métodos auxiliares

```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine
```

```python
# Função para plotar vetores
def plot_vectors(vectors, colors, alpha=1, figsize=(10,8)):
    plt.figure(figsize=figsize)
    plt.axvline(x=0, color='grey', lw=1)
    plt.axhline(y=0, color='grey', lw=1)

    for i in range(len(vectors)):
        x = np.concatenate([[0,0], vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],
                   alpha=alpha)
```

# Vetores

```python
# Definindo dois vetores
v1 = np.array([2, 5])
v2 = np.array([3, 1])

# Plotando vetores com função auxiliar
plot_vectors([v1, v2], ['blue', 'red'])
plt.xlim(-1, 5)
plt.ylim(-1, 6)
plt.show()
```

# Comprimento (Normas)

```python
norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)
print("Norma v1:", norm_v1)
print("Norma v2:", norm_v2)
```

# Produto Interno

```python
dot_product = np.dot(v1, v2)
print("Dot Product:", dot_product)
```

# Distância Euclidiana

```python
distance = np.linalg.norm(v1 - v2)
print("Distance entre v1 e v2 é:", distance)
```

# Ângulo

```python
# Cálculo de angulação com produto interno e normas
cos_theta = dot_product / (norm_v1 * norm_v2)
angle = np.arccos(cos_theta)
print("Angulo entre v1 e v2 (Radianos):", angle)
print("Angulo entre v1 e v2 (Graus):", np.degrees(angle))
```

# Ortogonalidade

```python
if dot_product == 0:
    print("v1 e v2 são ortogonais")
else:
    print("v1 and não são ortogonais")
```

# Exemplo de aplicação

```python
# Similaridade por cosseno
text_vector_1 = np.random.rand(100)
text_vector_2 = np.random.rand(100)

cos_similarity = 1 - cosine(text_vector_1, text_vector_2)
print("Similaridade por cosseno entre dois vetores:", cos_similarity)

plot_vectors([text_vector_1[:2], text_vector_2[:2]], ['green', 'orange'])
plt.show()
```
