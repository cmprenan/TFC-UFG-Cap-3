## **BASE ORTONORMAL**

---



## **Exemplo**

Para verificar que os vetores $v_1 = \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right)$ e $v_2 = \left(-\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right)$ formam uma base ortonormal, precisa-se verificar duas condições: ortogonalidade e normalização.

Para verificar a ortogonalidade, calculamos o produto interno entre $v_1$ e $v_2$:
    \begin{align*}
      v_1 \cdot v_2 &= \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right) + \left(-\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right) \\
      &= \frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}} +  \left(-\frac{1}{\sqrt{2}}\right) \cdot \frac{1}{\sqrt{2}} \\
      &= - \frac{1}{2} + \frac{1}{2} \\
      &= 0
    \end{align*}
   



```python

# Definindo os vetores v1 e v2
v1 = [1/math.sqrt(2), 1/math.sqrt(2)]
v2 = [1/math.sqrt(2), -1/math.sqrt(2)]

# Calculando o produto interno entre v1 e v2
produto_interno = round(sum(x * y for x, y in zip(v1, v2)))
print(produto_interno)
```

Plotando os vetores

```python
#Definindo os vetores
v1 = [1/math.sqrt(2), 1/math.sqrt(2)]
v2 = [1/math.sqrt(2), -1/math.sqrt(2)]


# Criando uma figura e eixos
fig, ax = plt.subplots()

# Adicionando os vetores ao gráfico
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')

# Configurando limites dos eixos
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

# Adicionando grade e legendas
ax.grid(True)
ax.legend()

# Mostrando o gráfico
plt.show()
```

Normalização:
Para verificar a normalização, calcula-se o módulo de cada vetor:
*      $\|v_1\| = \sqrt{\left(\frac{1}{\sqrt{2}}\right)^2 + \left(\frac{1}{\sqrt{2}}\right)^2} = \sqrt{\frac{1}{2} + \frac{1}{2}} = \sqrt{1} = 1$

*      $||v_2|| = \sqrt{\left(- \frac{1}{\sqrt{2}}\right)^2 + \left(\frac{1}{\sqrt{2}}\right)^2} = \sqrt{\frac{1}{2} + \frac{1}{2}} = \sqrt{1} = 1 $


Como ambos os vetores têm módulo igual a 1, eles estão normalizados.

```python
# Calculando o módulo de v1 e v2
modulo_v1 = round(math.sqrt(sum(x**2 for x in v1)))
modulo_v2 = round(math.sqrt(sum(x**2 for x in v2)))
print(modulo_v1)
print(modulo_v2)
```

##**Exemplo 02:**

 Encontre uma base ortonormal a partir da base:

$\beta = \{ (2,0,0),(0,2,2),(0,2,-2)\} $

```python
# Define os vetores da base beta
v1 = np.array([2, 0, 0])
v2 = np.array([0, 2, 2])
v3 = np.array([0, 2, -2])

# Calcula os vetores normalizados
u_n = v1 / np.linalg.norm(v1)
v_n = v2 / np.linalg.norm(v2)
w_n = v3 / np.linalg.norm(v3)

# Imprime os vetores normalizados
print("Vetor u_n:", u_n)
print("Vetor v_n:", v_n)
print("Vetor w_n:", w_n)

# Define a base ortonormal alpha
alpha = np.array([u_n, v_n, w_n])

# Imprime a base ortonormal alpha
print("\nBase ortonormal alpha:")
print(alpha)


# Calcula o tamanho dos vetores
tamanho_u_n = np.round(np.linalg.norm(u_n))
tamanho_v_n = np.round(np.linalg.norm(v_n))
tamanho_w_n = np.round(np.linalg.norm(w_n))

# Imprime o tamanho dos vetores normalizados
print("Tamanho de u_n:", tamanho_u_n)
print("Tamanho de v_n:", tamanho_v_n)
print("Tamanho de w_n:", tamanho_w_n)
```

##**COMPLEMENTO ORTOGONAL**

_______________________________________________________________________________

##**Exemplo**
Como encontrar o complemento ortogonal do seguinte espaço vetorial:
\[V = span{(1,2,3) , (2,3,4)}\]

O complemento ortogonal então $(x, y, z) \in V^{\perp}: $ \\
Utilizando a propriedade da ortogonalidade:\\
$\langle (x,y,z) , (1,2,3) \rangle = 0$\\
$\langle (x,y,z) , (2,3,4) \rangle = 0$\\
Montando a equação, e multipliacando a primeira por (-2):
\[
\begin{cases}
    x + 2y + 3z = 0 \cdot (-2) \\
    2x+3y+4z =  0
\end{cases}
\]
\\
$y = -2z$ substituindo na primeira equação: $ x + 2(-2z) + 3z = 0 $. Resultando:
\\
$x = z$
$(x, y, z) => (z,-2z, z) => z \cdot (1, -2, 1) $ \\

Logo o complemento ortogonal do espaço vetorial fica $ V^{\perp} = span \{(1,-2,1) \}$.

```python
# Definindo as variáveis
x, y, z = symbols('x y z')

# Definindo as equações
eq1 = Eq(x + 2*y + 3*z, 0)
eq2 = Eq(2*x + 3*y + 4*z, 0)

# Resolvendo o sistema de equações
sol = solve((eq1, eq2), (x, y, z))

# Imprimindo as soluções
print(sol)
```

## **Inner Product of Functions**

_____________________________________________________________________________



## **Exemplo**

Dadas as funções u e v, mostre a ortogonalidade.
* $\langle u,v \rangle = \int_{-\pi}^{\pi} sin(x)cos(x) \, dx$

Considerando:

$ C = sin(x) $ e
$dc = cos(x)dx $ \\
$ \langle u,v \rangle =  \int_{-\pi}^{\pi} Cdc  = [\frac{C^2}{2} ]_{-\pi}^{\pi} $
$ \langle u,v \rangle =  [\frac{sin(x)^2}{2} ]_{-\pi}^{\pi}  = 0 $


```python
# Definindo a variável x
x = symbols('x')

# Definindo as funções u e v
u = sin(x)
v = cos(x)

# Calculando o produto interno
produto_interno = integrate(u*v, (x, -(math.pi), (math.pi)))

# Imprimindo o valor do produto interno
print(f'O produto interno é: {produto_interno}')
```
